```python
df = df_final.copy()
```

```python
df.info()
```

```python
listaLikes = df['Likes'].tolist()
listaRet = df['Retweets'].tolist()
listaComents = df['Comentarios'].tolist()
```

```python
lista2Likes = []
lista2Ret = []
lista2Com = []
for i in range(len(listaLikes)):
    lista2Likes.append(str(listaLikes[i]))
for i in range(len(listaRet)):
    lista2Ret.append(str(listaRet[i]))
for i in range(len(listaComents)):
    lista2Com.append(str(listaComents[i]))
```

```python
df['Likes'] = pd.Series(lista2Likes)
df['Retweets'] = pd.Series(lista2Ret)
df['Comentarios'] = pd.Series(lista2Com)
```

```python
for i in range(len(df['Likes'])):
    if df['Likes'][i].find('mil') != -1:
        df['Likes'][i] = df['Likes'][i].replace(' mil',"")
        df['Likes'][i] = df['Likes'][i].replace(',',".")
        df['Likes'][i].rstrip()
        df['Likes'][i] = float(df['Likes'][i])
        df['Likes'][i] = (df['Likes'][i]*1000)

for i in range(len(df['Retweets'])):
    if df['Retweets'][i].find('mil') != -1:
        df['Retweets'][i] = df['Retweets'][i].replace(' mil',"")
        df['Retweets'][i] = df['Retweets'][i].replace(',',".")
        df['Retweets'][i].rstrip()
        df['Retweets'][i] = float(df['Retweets'][i])
        df['Retweets'][i] = (df['Retweets'][i]*1000)
        
for i in range(len(df['Comentarios'])):
    if df['Comentarios'][i].find('mil') != -1:
        df['Comentarios'][i] = df['Comentarios'][i].replace(' mil',"")
        df['Comentarios'][i] = df['Comentarios'][i].replace(',',".")
        df['Comentarios'][i].rstrip()
        df['Comentarios'][i] = float(df['Comentarios'][i])
        df['Comentarios'][i] = (df['Comentarios'][i]*1000)
```

```python
df.info()
```

```python
df.astype({'Likes': 'int','Retweets':'int', 'Comentarios':'int'}).dtypes
df = df.astype({'Likes': 'int','Retweets':'int', 'Comentarios':'int'})
```
